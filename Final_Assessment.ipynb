```python
#capping method
def iqr_capping(data1,variables):
  Q1=data1[variables].quantile(0.25)
  Q3=data1[variables].quantile(0.75)
  IQR=Q3-Q1
  upp_limit=Q3+(1.5*IQR)
  low_limit=Q1-(1.5*IQR)
  return low_limit,upp_limit
```

```python
low_applicant,upp_applicant=iqr_capping(data,'ApplicantIncome')
print('lower applicant=',low_applicant)
print('upper applicant=',upp_applicant)
```

```python
low_applicant,upp_applicant=iqr_capping(data,'ApplicantIncome')
print('lower applicant=',low_applicant)
print('upper applicant=',upp_applicant)
data['ApplicantIncome']=np.where(data['ApplicantIncome']>upp_applicant,upp_applicant,
                          np.where(data['ApplicantIncome']<low_applicant,low_applicant,data['ApplicantIncome']))
```

```python
low_coapplicant,upp_coapplicant=iqr_capping(data,'CoapplicantIncome')
print('lower coapplicant=',low_coapplicant)
print('upper coapplicant=',upp_coapplicant)
data['CoapplicantIncome']=np.where(data['CoapplicantIncome']>upp_coapplicant,upp_coapplicant,
                                 np.where(data['CoapplicantIncome']<low_coapplicant,low_coapplicant,data['CoapplicantIncome']))
```

```python
low_coapplicant,upp_coapplicant=iqr_capping(data,'LoanAmount')
print('lower coapplicant=',low_coapplicant)
print('upper coapplicant=',upp_coapplicant)
data['LoanAmount']=np.where(data['LoanAmount']>upp_coapplicant,upp_coapplicant,
                                 np.where(data['LoanAmount']<low_coapplicant,low_coapplicant,data['LoanAmount']))
```

```python
x1=data2.drop(['Gender','Married','Dependents','Education','Self_Employed','Loan_Amount_Term','Credit_History','Property_Area','Loan_Status'],axis=1)
```

```python
#minmax scaling

from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler(feature_range=(0,1))
x1=min_max_scaler.fit_transform(x1)
x1
```

```python
x1=pd.DataFrame(x1,columns=['ApplicantIncome','CoapplicantIncome','LoanAmount'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x3,y,random_state=42,test_size=0.25)
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(x_train,y_train)
y_pred=logit_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('Accuracy is',accuracy_score(y_test,y_pred))
print('Precisiom is',precision_score(y_test,y_pred,average='macro'))
print('Recall is',recall_score(y_test,y_pred,average='macro'))
print('F1 is',f1_score(y_test,y_pred,average='macro'))
confusion_matrix(y_test,y_pred)
```

```python
#KNN

from sklearn.neighbors import KNeighborsClassifier
metric_k=[]
neighbors=np.arange(3,15)


for k in neighbors:
  classifier=KNeighborsClassifier(n_neighbors=k,metric='manhattan')
  classifier.fit(x_train,y_train)
  y_pred=classifier.predict(x_test)
  acc=accuracy_score(y_test,y_pred)
  metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid
```

```python
classifier=KNeighborsClassifier(n_neighbors=14,metric='manhattan')
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
y_pred
```

```python
print('Accuracy is',accuracy_score(y_test,y_pred))
print('Precisiom is',precision_score(y_test,y_pred))
print('Recall is',recall_score(y_test,y_pred))
print('F1 is',f1_score(y_test,y_pred))
confusion_matrix(y_test,y_pred)
```

```python
#Decision tree

from sklearn.tree import DecisionTreeClassifier
dt_clf=DecisionTreeClassifier(random_state=42)
dt_clf.fit(x_train,y_train)
y_pred=dt_clf.predict(x_test)
```

```python
print('Accuracy is:',accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
#Random Forest

from sklearn.ensemble import RandomForestClassifier
rf_clf=RandomForestClassifier()
rf_clf.fit(x_train,y_train)
```

```python
y_pred=rf_clf.predict(x_test)
y_pred
```

```python
print('Accuracy is:',accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
#svm

from sklearn.svm import SVC
svmclf=SVC(kernel='rbf')
svmclf.fit(x_train,y_train)
y_pred=svmclf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy is:',accuracy_score(y_test,y_pred))
```

```python
#hyper parameter tuning

rf_clf=RandomForestClassifier(n_estimators=20,max_depth=3,criterion='entropy',min_samples_split=2)
rf_clf.fit(x_train,y_train)
```
